### Function to convert AVERAGE.ACCT.AGE and CREDIT.HISTORY.LENGTH to months 

```python
def duration(dur):
    yrs = int(dur.split(' ')[0].replace('yrs',''))
    mon = int(dur.split(' ')[1].replace('mon',''))
    return yrs*12+mon
```

```python
df['Avg_Loan_tenure'] = df['AVERAGE.ACCT.AGE'].apply(duration)
df['Time_since_1st_loan'] = df['CREDIT.HISTORY.LENGTH'].apply(duration)
```

```python
df.drop(['AVERAGE.ACCT.AGE','CREDIT.HISTORY.LENGTH'],axis=1,inplace=True)
```

### Age at the time of loan disbursal

```python
def age(dur):
    yr = int(dur.split('-')[2])
    if yr >=0 and yr<=19:
        return yr+2000
    else:
         return yr+1900

df['Date.of.Birth'] = df['Date.of.Birth'].apply(age)
df['DisbursalDate'] = df['DisbursalDate'].apply(age)
```

```python
df['Age_at_disbursal']=df['DisbursalDate']-df['Date.of.Birth']
```

```python
df.head()
```

```python
lower={}
upper={}
for i in numerical:
        q1=df[i].quantile(0.25)
        q3=df[i].quantile(0.75)
        iqr=q3-q1
        lower[i]=q1 - iqr * 1.5
        upper[i]=q3 + iqr * 1.5
```

```python
l=[]
for i in numerical:
    l.append(df.loc[(df[i]<lower[i])|(df[i]>upper[i]),i].index)
```
